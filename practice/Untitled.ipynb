### matplotlib workflow

```python
#  1. import matplotlib and get it ready
%matplotlib inline
import matplotlib.pyplot as plt

# 2 prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

#  3. set up plot
fig, ax = plt.subplots(figsize = (5,5) )

#  4. plot data
ax.plot(x, y)

# 5.  customize plot
ax.set(title = 'sample project', xlabel= 'x-axis', ylabel= 'y-axis')

#  6.  save the plot
fig.savefig('./sample-plot.png')
```

### making figures with numpy array

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
#  line plot
fix, ax = plt.subplots(figsize = (5,5) )
ax.plot(x, x**2)
```

```python
#  scatter plot
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(x, np.exp(x))
```

```python
# one more scatter
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(x, np.cos(x))
```

```python
# making a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter":8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut butter prices", xlabel= "Butters", ylabel="Prices $");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Nut butter prices", xlabel= "Prices $", ylabel="Butters");
```
